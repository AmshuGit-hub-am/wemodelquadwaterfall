```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#defining the quadratic function
def quadratic(x, a=-1, b=0, c=10):
  return a * x ** 2+ b * x + c
```

```python
#defining the stages of the waterfall model
stages =["Requirements","Design","Implementation","Testing","Deplyoment","Maintenence"]
x_pos= np.linspace(-2, 2, len(stages))
y_pos= quadratic(x_pos)

#plotting the quadratic curve
x_curve = np.linspace(-2.5, 2.5, 100)
y_curve = quadratic(x_curve)
plt.figure(figsize=(10,6))
plt.plot(x_curve, y_curve, color="#e73895", linestyle="--", label="Quadratic Curve")
for i in range(len(stages)):
  if i>0:
    #draw a vertical drop to simulate a waterfall
    plt.plot([x_pos[i-1], x_pos[i]], [y_pos[i-1], y_pos[i-1]], color="#8ca765", linewidth="2") #horizontal segment
    plt.plot([x_pos[i], x_pos[i]], [y_pos[i-1], y_pos[i]], color="#8ca765", linewidth="2") #vertical segment

    #place markers and labels
    plt.scatter(x_pos[i], y_pos[i], color="#6495ED", zorder=5)
    plt.text(x_pos[i], y_pos[i]+0.3, stages[i], ha="center", fontsize="10")

#styling the plot
plt.title("Waterfall Model of a Quadratic Function", fontsize=14)
plt.xlabel("Stages", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.axhline(0, color="black",linewidth="0.5",linestyle="--")
plt.axvline(0, color="black",linewidth="0.5",linestyle="--")
plt.grid(color="gray",linewidth=0.5, linestyle="--", alpha=0.7)
plt.legend()
plt.show()

```
